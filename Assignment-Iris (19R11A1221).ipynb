# Matplot library

```python
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11)#
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel("linspace array")
plt.ylabel("powers of array")
plt.title("grapg b/w x and y")
plt.show()
```

```python
plt.scatter(x,y,color = "purple")
plt.xlabel("linspace array")
plt.ylabel("powers of array")
plt.title("grapg b/w x and y")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(y,x,'r')
plt.subplot(2,2,2)
plt.scatter(x,y,color = 'orange')
plt.subplot(2,2,3)
plt.scatter(y,x,color = 'pink')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,0.8])
axes2 = fig.add_axes([0.1,0.1,1,0.8])
axes.plot(x,y,'b')
axes2.scatter(x,y,color = 'red')

```

```python
plt.bar(x,y,color = "yellow")
```

```python
plt.plot(x,y,'r',label = "xvsy")
plt.scatter(y,x,color = 'blue',label = "scatter x /y")
plt.legend()
```

```python
sns.distplot(iris["petal_length"])
```
