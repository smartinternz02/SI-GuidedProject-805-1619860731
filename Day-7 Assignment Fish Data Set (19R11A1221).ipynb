```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Fish.csv")
```

```python
data.isnull().any()
```

```python
x=data.iloc[:,[0,2,3,4,5,6]].values
y=data.iloc[:,1].values
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder#creates columns based on categories of textual data
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([("oh",OneHotEncoder(),[0])],remainder="passthrough")#add columns to first
x=ct.fit_transform(x)
```

```python
x
```

```python
data.Species.unique()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
ypred=mlr.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
accuracy=r2_score(ypred,y_test)
```

```python
accuracy
```
